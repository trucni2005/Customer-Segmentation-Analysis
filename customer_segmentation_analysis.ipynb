```python
Q1 = cleaned_data['Income'].quantile(0.25)
Q3 = cleaned_data['Income'].quantile(0.75)
IQR = Q3 - Q1
outliers = cleaned_data[(cleaned_data['Income'] < Q1 - 10 * IQR) | (cleaned_data['Income'] > Q3 + 10 * IQR)]
outliers
```

```python
Q1 = cleaned_data['Income'].quantile(0.25)
Q3 = cleaned_data['Income'].quantile(0.75)
IQR = Q3 - Q1
upper_bound = Q3 + 10 * IQR
lower_bound = Q1 - 10 * IQR

filtered_data = cleaned_data[(cleaned_data['Income'] >= lower_bound) & (cleaned_data['Income'] <= upper_bound)]
```

    <h2>Feature Engineering Summary</h2>

  <h3>Age Calculation</h3>
    <p>
      Calculated the age of customers by subtracting their year of birth from 2024.
    </p>

  <h3>Total Spending</h3>
    <p>
      Summed up the spending on various product categories, including wines, fruits, meat, fish, sweets, and gold products.
    </p>

  <h3>Children Count</h3>
    <p>
      Added the number of kids at home (<code>Kidhome</code>) and teenagers at home (<code>Teenhome</code>).
    </p>

  <h3>Marital Status Grouping</h3>
    <p>
      Consolidated marital statuses into two categories:
    </p>
    <ul>
      <li><strong>Partner:</strong> Includes "Married" and "Together".</li>
      <li><strong>Alone:</strong> Includes "Single", "Divorced", "Widow", "YOLO", and "Absurd".</li>
    </ul>

  <h3>Family Size</h3>
    <p>
      Calculated family size by combining the living arrangement (1 for "Alone", 2 for "Partner") and the number of children.
    </p>

  <h3>Parent Identification</h3>
    <p>
      Created a binary feature (<code>Is_Parent</code>) to indicate whether the customer has children (1 if they do, 0 otherwise).
    </p>
  
  <h3>Dropping some of the redundant features</h3>

```python
max_date = filtered_data['Dt_Customer'].max()
max_date
```

```python
filtered_data['Dt_Customer']
```

```python
filtered_data['Customer_For'] = (max_date - filtered_data['Dt_Customer']).dt.days
```

```python
filtered_data
```

```python
filtered_data["Marital_Status"].value_counts()
```

```python
filtered_data["Education"].value_counts()
```

```python
filtered_data_copy = filtered_data.copy()
```

```python
filtered_data_copy.loc[:, "Age"] = 2024 - filtered_data_copy["Year_Birth"]

filtered_data_copy.loc[:, "Spent"] = (
    filtered_data_copy["MntWines"] +
    filtered_data_copy["MntFruits"] +
    filtered_data_copy["MntMeatProducts"] +
    filtered_data_copy["MntFishProducts"] +
    filtered_data_copy["MntSweetProducts"] +
    filtered_data_copy["MntGoldProds"]
)

filtered_data_copy.loc[:, "Children"] = filtered_data_copy["Kidhome"] + filtered_data_copy["Teenhome"]

filtered_data_copy.loc[:, "Living_With"] = filtered_data_copy["Marital_Status"].replace({
    "Married": "Partner",
    "Together": "Partner",
    "Absurd": "Alone",
    "Widow": "Alone",
    "YOLO": "Alone",
    "Divorced": "Alone",
    "Single": "Alone"
})

filtered_data_copy.loc[:, "Family_Size"] = (
    filtered_data_copy.loc[:, "Living_With"].replace({"Alone": 1, "Partner": 2}) +
    filtered_data_copy.loc[:, "Children"]
)

filtered_data_copy["Is_Parent"] = np.where(filtered_data_copy["Children"] > 0, 1, 0)
```

```python
to_drop = ["Dt_Customer", "Z_CostContact", "Z_Revenue", "Year_Birth", "ID"]
filtered_data_copy = filtered_data_copy.drop(to_drop, axis=1)
```
